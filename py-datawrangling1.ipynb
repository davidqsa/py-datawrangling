```python
# Importación de librerías.
import numpy as np
import pandas as pd
import unidecode
```

Para realizar un análisis más rápido y sin complejizar el resultado, como consideración para esta evaluación, se extraerá, de cada archivo Excel del informe del banco central, 1 hoja del archivo colocaciones, 1 hoja del archivo de depositos y 1 hoja del archivo de inversiones. 

```python
# Lectura de cada uno de los archivos Excel de cada tipo de operación (colocación, captación, inversión).
# La fila donde comienza el archivo Excel es estándar. Así que definiremos la fila inicial.
fila_ini_excel = 6
# Extrae los datos de colocaciones, del archivo Excel.
df_coloc = pd.read_excel("sdbcol_092021.xlsx", sheet_name="1_01", header=fila_ini_excel-1)
# Extrae los datos de depósitos, del archivo Excel.
df_depos = pd.read_excel("sdbdep_092021.xlsx", sheet_name="2_01", header=fila_ini_excel-1)
# Extrae los datos de inversiones, del archivo Excel.
df_inver = pd.read_excel("sdbinv_092021.xlsx", sheet_name="3_01", header=fila_ini_excel-1)
```

Con lo anterior, se deduce que todos los DF terminan en la fila con índice 25, asi que podríamos cortar la muestra en esa fila.
Además, tenemos la primera columna sin encabezado ni valores en todos los DF. Esta columna también la eliminamos.

```python
# Limpio la extracción:
# elimino las ultimas filas de totales y comentarios, y
# elimino la primera columna extraída (columna vacía), para cada uno de los DF.
# Como se comprobó arriba, asumiremos que siempre el total de filas y columnas es el mismo para cada DF.
df_coloc = df_coloc.iloc[:26, 1:]
df_depos = df_depos.iloc[:26, 1:]
df_inver = df_inver.iloc[:26, 1:]
```

En la verificación, se detectan algunos valores 'ND', los cuales procederemos a reemplazar por 0 (cero). Esto ya que en el glosario
publicado por el banco central, indica que el valor 'ND' significa que no existen datos de ese banco para esa fecha,
por lo tanto, podemos indicar que en ese periodo el monto de la operación es 0 (cero).

```python
# Reemplazamos los valores 'ND'
df_coloc = df_coloc.replace("ND", 0)
df_depos = df_depos.replace("ND", 0)
df_inver = df_inver.replace("ND", 0)
# Mostramos para ver el cambio
df_coloc.tail(2)
```

Con lo anterior, comprobamos que los 3 DF tienen ahora la misma estructura.
Ahora lo que haremos es renombrar el encabezado de la primera columna en la nueva estructura de los DF, ya que
tiene un valor que no representa el contenido de dicha columna. Le pondremos "Banco", ya que el contenido
de esa columna es justamente el listado de los bancos participantes de la medición. 

```python
# Renombramos la primera columna de cada DF, con el valor "Banco".
# Especificamos el parámetro inplace para guardar el cambio.
df_coloc.rename(columns={df_coloc.columns[0]: 'Banco'}, inplace=True)
df_depos.rename(columns={df_depos.columns[0]: 'Banco'}, inplace=True)
df_inver.rename(columns={df_inver.columns[0]: 'Banco'}, inplace=True)
```

```python
# Verificamos el cambio
df_coloc.head(1)
```

Lo siguiente, es limpiar y mejorar los siguientes encabezados de columna, desde la 2da columna en adelante.
Estas columnas tienen como encabezado un valor "Fecha" de la forma 'yyyy-mm-dd 00:00:00'. Como cada título
se refiere a un mes de cada año, vamos a reducir el título al formato 'yyyy-mm', para dejarlo más limpio y legible.

```python
# Cambiar los nombres de encabezados de las columnas al formato "año-mes", para que se vean más limpios.
# Obtener los nombres de las columnas sin considerar la primera columna ("Banco").
columnas1 = df_coloc.columns[1:]
columnas2 = df_depos.columns[1:]
columnas3 = df_inver.columns[1:]
# Cambiar los nombres de los encabezados de las columnas al formato "año-mes"
nuevos_nombres1 = pd.to_datetime(columnas1, format="%Y-%m").strftime("%Y-%m")
nuevos_nombres2 = pd.to_datetime(columnas2, format="%Y-%m").strftime("%Y-%m")
nuevos_nombres3 = pd.to_datetime(columnas3, format="%Y-%m").strftime("%Y-%m")
# Actualiza encabezados para c/u de los DF.
df_coloc.columns = [df_coloc.columns[0]] + list(nuevos_nombres1)
df_depos.columns = [df_depos.columns[0]] + list(nuevos_nombres2)
df_inver.columns = [df_inver.columns[0]] + list(nuevos_nombres3)     
```

Ahora, verificamos que se hayan cambiado los encabezados.

```python
# Mostramos el cambio
df_coloc.head(1)   
```

```python
df_depos.head(1)
```

```python
df_inver.head(1)
```

Ahora tenemos la misma estructura en todos los DF. Para seguir, vamos a 'despivotear' cada uno de los DF.

```python
# Despivoteamos cada DF.
# La columna "Año-Mes" le pondremos "Anio-Mes" para no tener problemas con la 'ñ' en la exportación del .csv final.
df_coloc_fin = df_coloc.melt(id_vars='Banco', var_name='Anio-Mes', value_name='Monto')
df_depos_fin = df_depos.melt(id_vars='Banco', var_name='Anio-Mes', value_name='Monto')
df_inver_fin = df_inver.melt(id_vars='Banco', var_name='Anio-Mes', value_name='Monto')
# Mostramos un DF para validar el cambio.
df_coloc_fin.head(3)
```

Ahora, lo que tenemos que hacer es asignar una columna adicional a cada DF final, que indique a qué 'operación' bancaria corresponde
cada registro, ya que los datos vienen de planillas de operaciones distintas. Esto lo hacemos ANTES de juntar los DF, 
para tener diferenciadas _a priori_ las operaciones de colocación, depósitos e inversiones, en la misma tabla.

```python
df_coloc_fin['tipo_operacion'] = 'COLOCACIONES'
df_depos_fin['tipo_operacion'] = 'DEPOSITOS'
df_inver_fin['tipo_operacion'] = 'INVERSIONES'
# revisamos un DF para comprobar
df_inver_fin.head(10)
```

Luego, como ya tenemos 3 DF despivoteados finales, los vamos a concatenar para dejar un solo DF total.

```python
# concatenamos los 3 DF finales.
df_total = pd.concat([df_coloc_fin, df_depos_fin, df_inver_fin])
# mostramos el DF final
df_total
```

Ahora reindexamos los registros para ordenar un poco más el DF final.

```python
df_total.reset_index(inplace=True, drop=True)
# mostramos el DF.
df_total
```

Con esto, ya tenemos un DF final ordenado y limpio, con todos los datos de operaciones acumulados.
Podemos hacer algunas consultas para verificar la consistencia del DF.

```python
# Realizamos un "pivot" del DF final para demostrar que su info es consistente.
df_total.pivot_table(index='Banco', columns='tipo_operacion', values='Monto', aggfunc='sum', fill_value=0)
```

Por último quitamos todos los tildes de los nombres de bancos, para que el archivo .csv que se exportará, esté más limpio.

```python
# Función que reemplaza las letras con tildes, por la misma letra sin tilde, de la columna 'Banco'.
def quitar_tildes(texto):
    texto_sin_tildes = unidecode.unidecode(texto)
    return texto_sin_tildes

# Aplicar la función a los valores de la primera columna del DataFrame
df_total['Banco'] = df_total['Banco'].apply(quitar_tildes)
df_total.head(30)
```

Para finalizar, enviamos el DF resultante a un archivo .CSV con los datos finales.

```python
# Crea el archivo final
df_total.to_csv("total_operaciones_fin.csv", index=False)
```
